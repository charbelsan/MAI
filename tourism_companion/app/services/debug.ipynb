```python
import random
import pandas as pd

def get_random_circuit(datacsv, only_circuit=True, circuit_name=None):
    data = pd.read_csv(datacsv)
    list_circuits = data["circuit"].unique().tolist()
    if only_circuit and circuit_name == None:
        circuit = random.choice(list_circuits)
        print('random select circuit : ', circuit)
        names = data[data["circuit"] == circuit]["name"].tolist()
        longitudes = data[data["circuit"] == circuit]["longitude"].tolist()
        latitudes = data[data["circuit"] == circuit]["latitude"].tolist()
        resultats= "name"+str(names) + "lat" + str(latitudes) + "lon" + str(longitudes)
    elif only_circuit and circuit_name != None:
        print('select circuit by name : ', circuit_name)
        circuit = circuit_name
        names = data[data["circuit"] == circuit]["name"].tolist()
        longitudes = data[data["circuit"] == circuit]["longitude"].tolist()
        latitudes = data[data["circuit"] == circuit]["latitude"].tolist()
        resultats= "name"+str(names) + "lat" + str(latitudes) + "lon" + str(longitudes)
        
    else:
        print('select all circuits')
        resultats = ""
        for circuit in list_circuits:
            names = data[data["circuit"] == circuit]["name"].tolist()
            longitudes = data[data["circuit"] == circuit]["longitude"].tolist()
            latitudes = data[data["circuit"] == circuit]["latitude"].tolist()
            resultats += "name"+str(names) + "lat" + str(latitudes) + "lon" + str(longitudes) + "#"
    
    return resultats

datacsv = "/home/boubacar-diallo/LLAMA/Challenge/MAI1/tourism_companion/benin_tourist_circuits.csv"
resultats = get_random_circuit(datacsv, only_circuit=False, circuit_name="Coastal and Beach Relaxation")

resultats
```

```python
import base64

def wav_to_base64(file_path):
    # Read the audio file as binary
    with open(file_path, 'rb') as f:
        audio_content = f.read()

    # Convert audio content to Base64
    base64_audio = base64.b64encode(audio_content).decode('utf-8')

    return base64_audio

def save_base64_to_file(base64_data, file_path):
    # Write the Base64 data to a text file
    with open(file_path, 'w') as f:
        f.write(base64_data)

# Example usage
audio_file_path = '/home/boubacar-diallo/LLAMA/Challenge/MAI1/tourism_companion/temp.wav'
output_file_path = '/home/boubacar-diallo/LLAMA/Challenge/MAI1/tourism_companion/audio_base64.txt'

# Convert audio file to Base64
base64_audio_data = wav_to_base64(audio_file_path)

audio_bytes = base64.b64decode(base64_audio_data)
print(audio_bytes)

# Save Base64 data to text file
save_base64_to_file(base64_audio_data, output_file_path)

print(f"Base64 audio saved to {output_file_path}")
```
